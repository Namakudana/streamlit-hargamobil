# Jumlah mobil berdasarkan tahun

```python
years = oto.groupby('year').count()[['tax']].sort_values(by='tax',ascending=True).reset_index()
years = years.rename(columns={'tax':'numberOfCars'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=years['year'], y=years['numberOfCars'], color='royalblue')
plt.xticks(rotation=60)
```

# Jumlah mobil berdasarkan model

```python
models = oto.groupby('model').count()[['tax']].sort_values(by='tax',ascending=True).reset_index()
models = models.rename(columns={'tax':'numberOfCars'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=models['model'], y=models['numberOfCars'], color='royalblue')
plt.xticks(rotation=60)
```

# Ukuran mesin

```python
engine = oto.groupby('cc').count()[['tax']].sort_values(by='tax').reset_index()
engine = engine.rename(columns={'tax':'count'})
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x=engine['cc'], y=engine['count'], color='royalblue')
```

# Seleksi Fitur

```python
features = ['year', 'mileage', 'tax', 'cc']
x = oto[features]
y = oto['price']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# Membuat model regresi linear

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

# Membuat inputan model regresi linear

```python
#year=2019, mileage=80.000, tax=29000000, cc=1198
input_data = np.array([[2019,120000,1000000,1198]])

prediction = lr.predict(input_data)
print('Estimasi harga mobil dalam IDR :', prediction)
```
